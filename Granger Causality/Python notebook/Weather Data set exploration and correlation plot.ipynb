# Date formatting

```python
def add_date_columns(df):
    date_columns = ['Formatted Date']
    for column in date_columns:
        df[column] = pd.to_datetime(df[column], format='%Y/%m/%d %H:%M:%S.%f')
        df[column+'.year'] = df[column].dt.year
        df[column+'.month'] = df[column].dt.month
        df[column+'.dayofmonth'] = df[column].dt.day
        df[column+'.Hour'] = df[column].dt.hour
        df[column+'.Minutes'] = df[column].dt.minute
        df[column+'.Seconds'] = df[column].dt.second
    return df
                                    
def preprocess(df):
    df = add_date_columns(df)
    return df
    
df=preprocess(df)
# print("head", df.head())
# print("describe", df.describe())
# Since the seconds and minutes are zero, we can drop the information.
df = df.drop(['Formatted Date','Formatted Date.Minutes', 'Formatted Date.Seconds'],axis=1)
```

# Categorical Encoding

```python
from sklearn.preprocessing import LabelEncoder
def encode(df):
    for col in df.columns:
        if df[col].dtype == "object":
            le = LabelEncoder()
            df_list = list(df[col].values.astype(str))
            le.fit(df_list)
            df[col] = le.transform(df_list)
    return df

df=encode(df)
```

```python
df.head()
```

```python
list_columns=df.columns
len(list_columns)
list_all=[]
for i in list_columns:
    list_all.append(df[i].values)
```

```python
import scipy.stats as sc
P = [[0 for i in range(len(list_columns))] for j in range(len(list_columns))]
S = [[0 for i in range(len(list_columns))] for j in range(len(list_columns))]
K = [[0 for i in range(len(list_columns))] for j in range(len(list_columns))]
for i in range(len(list_columns)):
    for j in range(i,len(list_columns)):
        P[i][j]=sc.pearsonr(list_all[i],list_all[j])[0]  
        S[i][j]=sc.spearmanr(list_all[i],list_all[j])[0]
        K[i][j]=sc.kendalltau(list_all[i],list_all[j])[0]
        P[j][i]= P[i][j]
        S[j][i]= S[i][j]
        K[j][i]= K[i][j]
        print("Pearson correlation for",list_columns[i], "and",list_columns[j], "is ", P[i][j])
```
