# Generation of the correlation matrix

```python
import pandas as pd
import scipy.stats as sc
df=pd.read_csv('data_vertical.csv')
list_all=[df['Sensor1'].values,df['Sensor2'].values,df['Sensor3'].values,df['Sensor4'].values,df['Sensor5'].values ]
# data_vertical.csv contains the values for the five sensors
# list_all = [sensor1,sensor2,sensor3,sensor4,sensor5]
P = [[0 for i in range(5)] for j in range(5)]
S = [[0 for i in range(5)] for j in range(5)]
K = [[0 for i in range(5)] for j in range(5)]
for i in range(5):
    for j in range(i,5):
        P[i][j]=sc.pearsonr(list_all[i],list_all[j])[0]  
        S[i][j]=sc.spearmanr(list_all[i],list_all[j])[0]
        K[i][j]=sc.kendalltau(list_all[i],list_all[j])[0]
        P[j][i]= P[i][j]
        S[j][i]= S[i][j]
        K[j][i]= K[i][j]
        print("Pearson correlation for sensor", i+1, "and sensor", j+1, "is ", P[i][j])
```

# Function to take a correlation matrix value and process them using fixed fuzzy system and then computing the matrix.

```python
def fun(correlation,membership,values):
    '''
    correlation matrix n*n symmetric matrix
    So its better to compute for half of the values. 
    In rectangle due to symmetry the sum of the y values is equal to 1.
    membership is nothing but similar to the membermembership function--- [x1 x2 x3 x4]
    Right now, I'm defining only between two things. 
    y=0 and  y=1 for some x1 and x2
    y=1 and  y=0 for the same x1 and x2 
    I have to define four lines for two different ranges. 
    So taking the arguments in membermembership function as (x1,x2) for the intersection of the Weak and Medium
    (x3,x4) for the intersection of Medium and Strong. 
    It should output the values Weak*possibility_Weak +  Medium*possibility_Medium + Strong*possibility_Strong
    values---[Weak, Medium, Strong]
    '''
    K=[[0 for j in range(len(correlation))] for i in range(len(correlation))]
    for i in range(len(correlation)):
        for j in range(i,len(correlation)):
            Possibility=[0,0,0]
            if correlation[i][j] <= membership[0] :
                Possibility[0]=1
            elif (correlation[i][j] > membership[0]) and (correlation[i][j] < membership[1]):
                Possibility[0]=(correlation[i][j])/(membership[0]- membership[1]) - (membership[1])/(membership[0]- membership[1])
                Possibility[1] = 1- Possibility[0]
                #Possibility[1]=-( (correlation[i][j])/(membership[0]- membership[1]) - (membership[0])/(membership[0]- membership[1]))
            elif (correlation[i][j] >= membership[1]) and (correlation[i][j] <= membership[2]):
                Possibility[1]=1
            elif (correlation[i][j] > membership[2]) and (correlation[i][j] < membership[3]):
                Possibility[1]=(correlation[i][j])/(membership[2]- membership[3]) - (membership[3])/(membership[2]- membership[3])
                Possibility[2]=1- Possibility[1]
            elif (correlation[i][j] >= membership[3]) and (correlation[i][j] <= 1):
                Possibility[2]= 1
            else:
                print("Invalid Input")
                break
            total=0
            for k in range(len(Possibility)):
                total+=Possibility[k]*values[k]
            K[i][j]=total
            K[j][i]=total
    return K
membership=[0.88,0.895,0.9,0.95]
# membership=[0.2,0.3,0.8,0.86]
values=[0,1,2]
print("Pearson correlation coefficient")
print(fun(P,membership,values))
print("Spearman correlation coefficient")
print(fun(S,membership,values))
print("Kendallâ€™s Tau correlation coefficient")
print(fun(K,membership,values))
```
